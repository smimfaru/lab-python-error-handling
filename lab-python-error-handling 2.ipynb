````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory
````

````python
def calculate_total_price(products):
    prices = {}
    for product in products:
        valid_price = False
        while not valid_price:
            try:
                price = float(input(f"Enter the price of {product}: "))
                if price < 0:
                    raise ValueError("Invalid price! Please enter a non-negative value.")
                valid_price = True
            except ValueError as error:
                print(f"Error: {error}")
        prices[product] = price
    return prices
````

````python
def get_customer_orders(inventory):
    orders = {}
    while True:
        try:
            num_orders = int(input("Enter the number of different products to order: "))
            if num_orders < 0:
                raise ValueError("Invalid number! Please enter a non-negative value.")
            break
        except ValueError as error:
            print(f"Error: {error}")

    for _ in range(num_orders):
        while True:
            product_name = input("Enter the product name: ")
            if product_name not in inventory:
                print("Error: This product is not in the inventory. Please enter a valid product name.")
                continue

            if inventory[product_name] <= 0:
                print(f"Error: {product_name} is out of stock.")
                continue

            try:
                quantity = int(input(f"Enter the quantity for {product_name}: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                elif quantity > inventory[product_name]:
                    print(f"Error: Only {inventory[product_name]} {product_name}s are available.")
                else:
                    orders[product_name] = quantity
                    inventory[product_name] -= quantity
                    break
            except ValueError as error:
                print(f"Error: {error}")

    return orders
````

````python
products = ["apple", "banana", "orange"]

# Step 1: Initialize inventory with error handling
inventory = products

# Step 2: Set prices with error handling
prices = calculate_total_price(products)

# Step 3: Get customer orders with error handling
orders = get_customer_orders(inventory)

# Display final order summary
print("\nOrder Summary:")
total_cost = 0
for product, quantity in orders.items():
    cost = prices[product] * quantity
    total_cost += cost
    print(f"{quantity} x {product} @ {prices[product]} each = {cost}")

print(f"Total Cost: {total_cost}")
````
